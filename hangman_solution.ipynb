```python
import random


from word_logo.hangman_logo import stages, logo 
from word_logo.hangman_words import word_list

lives = 3

print(logo)

chosen_word = random.choice(word_list)
#print(chosen_word)

placeholder = ""
word_length = len(chosen_word)
for position in range(word_length):
    placeholder += "-"

print(f'\n')
print(f'''It's a {len(placeholder)} letter word to guess: ''' + placeholder)
print(f'\n')

game_over = False
correct_letters = []

while not game_over:

    print(f"****************************{lives}/3 LIVES LEFT****************************")
    print(f'\n')
    
    guess = input("Guess a letter: ").lower()

    if guess in correct_letters:
        print(f"You've already guessed {guess}")

    display = ""

    for letter in chosen_word:
        if letter == guess:
            print(f"That's correct you guessed {guess}, that's in the word. ")
            display += letter
            correct_letters.append(guess)
        elif letter in correct_letters:
            display += letter
        else:
            display += "-"

   

    if guess not in chosen_word:
        lives -= 1
        print(f"You guessed {guess}, that's not in the word. You lose a life.")

        if lives == 0:
            game_over = True
            print(f'\n')
            print(f"***********************IT WAS {chosen_word}! YOU LOSE**********************")
    
    print(f'\n')
    print("Word to guess {len(placeholder)}letters : " + display)

    if "-" not in display:
        game_over = True
        print(f'\n')
        print("****************************YOU WON****************************")

    print(stages[lives])
```
